```python
import pandas as pd

df = pd.read_csv("pizzas.csv")
df.head(10)
```

```python
df.plot(kind="scatter", x="diametro", y="preço")
```

```python
from sklearn import linear_model

model = linear_model.LinearRegression()
x = df[["diametro"]]
y = df[["preço"]]
model.fit(x, y)
```

```python
print(model.predict([[30]])[0][0])
```
